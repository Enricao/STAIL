## Dataset CelebAMask-HQ

The dataset consist of 30000 images. 41 attributes are available. The attributes convey information about the features of the person in a given image (e. g. whether the person is considered to be attractive or not, whether the person has a big nose or not, . . .) The values are always -1 and 1 representing NO and YES answers respectively.

There is another text file available called CelebAMask-HQ-pose-anno. The file conveys information about the position of the face.

```python
def attribute_function():
    import matplotlib.pyplot as plt
    %matplotlib inline
    print('List of attributes:\n',data.columns.values[1:],'\n')
    attribute = input('Attribute: ')
    
    if data[attribute].value_counts().index[0] >= data[attribute].value_counts().index[1]:
        y = np.array([data[attribute].value_counts().values[0],
                     data[attribute].value_counts().values[1]])
        #lbl = [round(data[attribute].value_counts().values[0]/data.shape[0],2),
        #       round(data[attribute].value_counts().values[1]/data.shape[0],2)]
    else:
        y = np.array([data[attribute].value_counts().values[1],
                     data[attribute].value_counts().values[0]])
        #lbl = [round(data[attribute].value_counts().values[1]/data.shape[0],2),
        #       round(data[attribute].value_counts().values[0]/data.shape[0],2)]
    
    plt.figure(figsize=(6,6))
    #plt.subplot(8, 5, 2) 
    plt.pie(y, 
            startangle = 90,
            #labels = lbl,
            colors=['green','red'],
            autopct='%1.0f%%')
    circle = plt.Circle( (0,0), 0.35, color='white')
    p=plt.gcf()
    p.gca().add_artist(circle)
    plt.legend(['YES', 'NO'])
    plt.title(attribute)
    plt.show(block=False)

    print(pd.DataFrame({'Count': y},
                      index=['YES','NO']),'\n')
```

```python
# CHOOSE THE ATTRIBUTE (the dunut chart with percentage and frequency table with absolute values will be shown)
attribute_function()
```
